```python
import pandas as pd
from surprise import Reader, Dataset, SVD
from surprise.model_selection import train_test_split


reader = Reader(rating_scale=(1, 5))

data = Dataset.load_from_df(df[['user_id_maped', 'program_name', 'rating']], reader)


trainset, testset = train_test_split(data, test_size=0.2)

algo = SVD()
algo.fit(trainset)

# Get recommendations for a specific user
user_id = 26138
program_names = df['program_name'].unique()
program_ids = [trainset.to_inner_iid(program_name) for program_name in program_names if program_name in trainset._raw2inner_id_items]
user_ratings = [algo.predict(user_id, program_id).est for program_id in program_ids]
program_ratings = list(zip(program_names, user_ratings))
top_recommendations = sorted(program_ratings, key=lambda x: x[1], reverse=True)[:5]

# Print the top recommendations
print("Top Recommendations for User", user_id)
for recommendation in top_recommendations:
    program_name = recommendation[0]
    predicted_rating = recommendation[1]
    print(f"Program: {program_name}, Predicted Rating: {predicted_rating}")
```

```python
import pandas as pd
from surprise import Reader, Dataset, SVD
from surprise.model_selection import train_test_split


# Create the Surprise reader and dataset
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(df[['user_id_maped', 'program_name', 'rating']], reader)

# Split the data into training and testing sets
trainset, testset = train_test_split(data, test_size=0.2)

# Train the SVD algorithm
algo = SVD()
algo.fit(trainset)

# Find the program ID for "Moana"
moana_program_id = trainset.to_inner_iid("Moana")

# Find users who watched "Moana"
users_who_watched_moana = [uid for (uid, iid, rating) in trainset.all_ratings() if iid == moana_program_id]

# Generate recommendations for each user who watched "Moana"
recommendations = []
for user_id in users_who_watched_moana:
    # Get the list of program IDs for which the user has not provided ratings
    program_ids = [iid for iid in trainset.all_items() if (user_id, iid) not in trainset.ur]

    # Predict ratings for the programs and sort them in descending order
    program_ratings = [(trainset.to_raw_iid(iid), algo.predict(user_id, iid).est) for iid in program_ids]
    top_recommendations = sorted(program_ratings, key=lambda x: x[1], reverse=True)[:5]

    # Add the recommendations to the list
    recommendations.extend(top_recommendations)

# Sort the recommendations by predicted rating in descending order
recommendations = sorted(recommendations, key=lambda x: x[1], reverse=True)

# Print the top 5 recommendations for people who watched "Moana"
print("Top Recommendations for People who watched 'Moana':")
for recommendation in recommendations[:5]:
    program_id = recommendation[0]
    predicted_rating = recommendation[1]
    program_info = df[df['program_name'] == program_id][['program_name', 'program_genre']].iloc[0]
    program_name = program_info['program_name']
    program_genre = program_info['program_genre']
    print(f"Program: {program_name}, Genre: {program_genre}")
```
